```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.express as px
```

```python
base_sinasc = pd.read_csv('./SINASC_RO_2019.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
base_sinasc
```

```python
for i in ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']:
    sns.countplot(x=i, data=base_sinasc, palette='pastel')
    plt.xlabel(i)
    plt.ylabel('Contagem')
    plt.title(f"{i} - Gráfico de Barras de Contagem")
    plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
media_APGAR5_consultas = base_sinasc.groupby('CONSULTAS')['APGAR5'].mean()
sns.barplot(x=media_APGAR5_consultas.index, y=media_APGAR5_consultas.values, palette='pastel')
plt.xlabel('Consultas')
plt.ylabel('Média por APGAR5')
plt.title('APGAR5 médio por consultas')
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

for i, variavel in enumerate(['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']):
    sns.countplot(x=variavel, data=base_sinasc, ax=axs[i//2,i%2], palette='pastel')
    sns.set(font_scale=1.2)
    axs[i//2,i%2].set_xlabel(variavel)
    axs[i//2,i%2].set_ylabel('Contagem')
    axs[i//2,i%2].set_title(f'Gráfico de Barra de Contagem {variavel}')
    plt.subplots_adjust(wspace=0.3, hspace=0.5)
```
