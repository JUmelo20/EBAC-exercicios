```python
df.columns
```

```python
df['log_renda'] = np.log(df['renda'])
```

```python
y, X = dmatrices('log_renda ~ sexo + posse_de_veiculo + posse_de_imovel + qtd_filhos + tipo_renda + educacao + estado_civil + tipo_residencia + idade + tempo_emprego + qt_pessoas_residencia + mau',
                 data=df, return_type='dataframe')

model = sm.OLS(y, X)
result = model.fit()
```

```python
result.summary()
```

```python
y, X = dmatrices('log_renda ~ posse_de_veiculo + qtd_filhos + tipo_renda + educacao + estado_civil + tipo_residencia + qt_pessoas_residencia + mau',
                 data=df, return_type='dataframe')

model = sm.OLS(y, X)
result = model.fit()
```

```python
result.summary()
```
